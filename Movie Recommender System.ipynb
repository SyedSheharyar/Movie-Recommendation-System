```python
ps = PorterStemmer()
```

```python
def stem(txt):
    lst = []
    
    for i in txt.split():
        lst.append(ps.stem(i))
        
    return " ".join(lst)
```

```python
# recommendation
```

```python
final_df.head(2)
```

```python
movie_index = final_df[final_df["title"] == "Avatar"].index[0]
movie_index
```

```python
distances = similarity[movie_index]
distances
```

```python
index_lst = list(enumerate(distances))
index_lst[:10]
```

```python
similar_movie = sorted(index_lst, reverse = True, key = lambda x:x[1])[1:11]
similar_movie
```

```python
for i in similar_movie:
    print(final_df.iloc[i[0],1])
```

```python
def recommender(movie_name):
    movie_index = final_df[final_df["title"] == movie_name].index[0]
    distances = similarity[movie_index]
    index_lst = list(enumerate(distances))
    similar_movie = sorted(index_lst, reverse = True, key = lambda x:x[1])[1:11]
    for i in similar_movie:
        print(final_df.iloc[i[0],1])
```

```python
recommender("Batman")
```

```python
recommender("Thor")
```
